```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (randomly generated)
data = np.random.randn(1000)  # Generating 1000 random data points from a standard normal distribution

# Create a histogram
plt.hist(data, bins=30, edgecolor='black')  # 'bins' parameter determines the number of bins in the histogram

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")

# Show the plot
plt.show()
```

```python
# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d 
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d') 
 
# defining all 3 axis
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot geeks for geeks')
plt.show() 
```
