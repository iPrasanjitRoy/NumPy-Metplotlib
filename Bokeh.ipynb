```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers

output_file('Flowers.html')

p = figure(title = "Flowers")
p.xaxis.axis_label = "paltel lenght"
p.yaxis.axis_label = "paltel widhth"
p.circle(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

# Create a figure
p = figure(title="Scatter Plot Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a scatter plot
p.scatter(x, y, size=10, color="blue", legend_label="Data Points")

# Show the legend
p.legend.location = "top_left"

# Show the plot in the notebook
output_notebook()
show(p)
```

```python
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title = "line graph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.line(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]
output_file("line.html")
p = figure(title = "line plot")
p.line(x,y)
show(p)
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]
output_file("line.html")
p = figure(title = "line plot")
p.scatter(x,y,fill_color = 'red' , legend_label = 'red points',size = 20)
show(p)
```
